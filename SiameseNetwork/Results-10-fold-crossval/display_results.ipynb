```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
sns.set_theme()
```

```python
path_xite_all = "xite_all.csv"
path_xite_all_random_forest = "xite_all_random_forest.csv"
path_xite_electric = "xite_electric.csv"
path_xite_heat = "xite_heat.csv"
path_xite_tonic = "xite_tonic.csv"
path_xite_phasic = "xite_phasic.csv"
path_xite_tonic_heat = "xite_tonic_heat.csv"
path_xite_tonic_electric = "xite_tonic_electric.csv"
path_xite_phasic_heat = "xite_phasic_heat.csv"
path_xite_phasic_electric = "xite_phasic_electric.csv"
```

```python
xite_all = pd.read_csv(path_xite_all)
xite_all_random_forest = pd.read_csv(path_xite_all_random_forest)
xite_electric = pd.read_csv(path_xite_electric)
xite_heat = pd.read_csv(path_xite_heat)

xite_phasic = pd.read_csv(path_xite_phasic)
xite_phasic_heat = pd.read_csv(path_xite_phasic_heat)
xite_phasic_electric = pd.read_csv(path_xite_phasic_electric)
xite_tonic = pd.read_csv(path_xite_tonic)
xite_tonic_heat = pd.read_csv(path_xite_tonic_heat)
xite_tonic_electric = pd.read_csv(path_xite_tonic_electric)
```

```python
df = pd.DataFrame([])
df = pd.concat([df, pd.DataFrame([xite_all["accuracy"].mean(), xite_all["recall"].mean(), xite_all["precision"].mean(), xite_all["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite all"})])
df = pd.concat([df, pd.DataFrame([xite_all_random_forest["accuracy"].mean(), xite_all_random_forest["recall"].mean(), xite_all_random_forest["precision"].mean(), xite_all_random_forest["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite all random forest"})])
df = pd.concat([df, pd.DataFrame([xite_heat["accuracy"].mean(), xite_heat["recall"].mean(), xite_heat["precision"].mean(), xite_heat["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite heat"})])
df = pd.concat([df, pd.DataFrame([xite_electric["accuracy"].mean(), xite_electric["recall"].mean(), xite_electric["precision"].mean(), xite_electric["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite electric"})])
df = pd.concat([df, pd.DataFrame([xite_phasic["accuracy"].mean(), xite_phasic["recall"].mean(), xite_phasic["precision"].mean(), xite_phasic["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite phasic"})])
df = pd.concat([df, pd.DataFrame([xite_phasic_heat["accuracy"].mean(), xite_phasic_heat["recall"].mean(), xite_phasic_heat["precision"].mean(), xite_phasic_heat["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite phasic heat"})])
df = pd.concat([df, pd.DataFrame([xite_phasic_electric["accuracy"].mean(), xite_phasic_electric["recall"].mean(), xite_phasic_electric["precision"].mean(), xite_phasic_electric["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite phasic electric"})])
df = pd.concat([df, pd.DataFrame([xite_tonic["accuracy"].mean(), xite_tonic["recall"].mean(), xite_tonic["precision"].mean(), xite_tonic["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite tonic"})])
df = pd.concat([df, pd.DataFrame([xite_tonic_heat["accuracy"].mean(), xite_tonic_heat["recall"].mean(), xite_tonic_heat["precision"].mean(), xite_tonic_heat["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite tonic heat"})])
df = pd.concat([df, pd.DataFrame([xite_tonic_electric["accuracy"].mean(), xite_tonic_electric["recall"].mean(), xite_tonic_electric["precision"].mean(), xite_tonic_electric["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite tonic electric"})])
df
```

```python
xite_all["accuracy"].plot.kde()
xite_electric["accuracy"].plot.kde()
xite_heat["accuracy"].plot.kde()
plt.legend(["all", "electric", "heat"])
plt.title("Accuracy Classification XITE")
plt.show()
```
