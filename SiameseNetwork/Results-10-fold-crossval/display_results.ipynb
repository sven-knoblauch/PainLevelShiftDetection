```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
sns.set_theme()
```

```python
path_xite_all = "xite_all_distance_0.csv"
path_xite_electric = "xite_electric_distance_0.csv"
path_xite_heat = "xite_heat_distance_0.csv"
```

```python
xite_all = pd.read_csv(path_xite_all)
xite_electric = pd.read_csv(path_xite_electric)
xite_heat = pd.read_csv(path_xite_heat)
```

```python
df = pd.DataFrame([])
df = pd.concat([df, pd.DataFrame([xite_all["accuracy"].mean(), xite_all["recall"].mean(), xite_all["precision"].mean(), xite_all["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite all"})])
df = pd.concat([df, pd.DataFrame([xite_electric["accuracy"].mean(), xite_electric["recall"].mean(), xite_electric["precision"].mean(), xite_electric["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite electric"})])
df = pd.concat([df, pd.DataFrame([xite_heat["accuracy"].mean(), xite_heat["recall"].mean(), xite_heat["precision"].mean(), xite_heat["f1"].mean()]).T.rename(columns={0: "acc", 1: "recall", 2: "precision", 3: "f1"}, index={0: "xite heat"})])
df
```

```python
xite_all["accuracy"].plot.kde()
xite_electric["accuracy"].plot.kde()
plt.legend(["all", "electric"])
plt.title("Accuracy Classification XITE")
plt.show()
```
