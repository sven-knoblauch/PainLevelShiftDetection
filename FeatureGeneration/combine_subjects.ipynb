```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import os
```

### Get path and all subjects

```python
path = "features\INTENSE2\\"
subjects = [x[:-4] for x in os.listdir(path)]
```

### Load all pickle files into one DataFrame

```python
#only for intense with 3 class problem

nan = [18,  19,  20,  63,  64,  65,  81,  82,  83, 129, 130, 131, 132, 133, 134, 642, 643, 644, 714, 715, 716, 873, 874, 875, 900, 901, 902]
```

```python
data = pd.read_pickle(path+subjects[0]+".pkl")
for subject in subjects[1:]:
    tmp = pd.read_pickle(path + subject + ".pkl")
    data = data.append(tmp, ignore_index=True)


#remove nan rows (7 rows in total)
#data.dropna(inplace=True)
data.drop(index=nan, inplace=True)
```

### standardize each subject

```python
#label = data["Condition"]
subjects = data["subject"]
pain = data["pain"]
data = data.drop(["pain"], axis=1)
```

```python
x = data.groupby("subject")
data = data.drop(["subject"], axis=1)
mean, std = x.transform("mean"), x.transform("std")
normalized = (data - mean) / std
```

```python
#normalized["label"] = label
normalized["subject"] = subjects
normalized["pain"] = pain
```

```python
normalized
```

```python
normalized = normalized.reset_index()
normalized = normalized.drop(["index"], axis=1)
```
