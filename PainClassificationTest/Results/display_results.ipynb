```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
sns.set_theme()
```

# LOSO

```python
path_xite_all = "xite_all.csv"
path_xite_all_embedding = "xite_all_embedding.csv"
path_intense = "intense.csv"
path_intense_embedding = "intense_embedding.csv"
path_xite_electric = "xite_electric.csv"
path_xite_heat = "xite_heat.csv"
path_xite_tonic = "xite_tonic.csv"
path_xite_tonic_electric = "xite_tonic_electric.csv"
path_xite_tonic_heat = "xite_tonic_heat.csv"
path_xite_phasic = "xite_phasic.csv"
path_xite_phasic_electric = "xite_phasic_electric.csv"
path_xite_phasic_heat = "xite_phasic_heat.csv"
path_xite_strong = "xite_strong.csv"
path_xite_strong_heat = "xite_strong_heat.csv"
path_xite_strong_electric = "xite_strong_electric.csv"
```

```python
xite_all = pd.read_csv(path_xite_all)
xite_all_embedding = pd.read_csv(path_xite_all_embedding)
intense = pd.read_csv(path_intense)
intense_embedding = pd.read_csv(path_intense_embedding)
xite_electric = pd.read_csv(path_xite_electric)
xite_heat = pd.read_csv(path_xite_heat)
xite_tonic = pd.read_csv(path_xite_tonic)
xite_tonic_electric = pd.read_csv(path_xite_tonic_electric)
xite_tonic_heat = pd.read_csv(path_xite_tonic_heat)
xite_phasic = pd.read_csv(path_xite_phasic)
xite_phasic_electric = pd.read_csv(path_xite_phasic_electric)
xite_phasic_heat = pd.read_csv(path_xite_phasic_heat)
xite_strong = pd.read_csv(path_xite_strong)
xite_strong_heat = pd.read_csv(path_xite_strong_heat)
xite_strong_electric = pd.read_csv(path_xite_strong_electric)
```

```python
df = pd.DataFrame([])
df = pd.concat([df, pd.DataFrame([intense["accuracy"].mean(), intense["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "intense"})])
df = pd.concat([df, pd.DataFrame([intense_embedding["accuracy"].mean(), intense_embedding["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "intense embedding"})])
df = pd.concat([df, pd.DataFrame([xite_all_embedding["accuracy"].mean(), xite_all_embedding["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite all embedding"})])
df = pd.concat([df, pd.DataFrame([xite_all["accuracy"].mean(), xite_all["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite all"})])
df = pd.concat([df, pd.DataFrame([xite_electric["accuracy"].mean(), xite_electric["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_electric"})])
df = pd.concat([df, pd.DataFrame([xite_heat["accuracy"].mean(), xite_heat["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_heat"})])
df = pd.concat([df, pd.DataFrame([xite_phasic["accuracy"].mean(), xite_phasic["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_phasic"})])
df = pd.concat([df, pd.DataFrame([xite_tonic["accuracy"].mean(), xite_tonic["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_tonic"})])
df = pd.concat([df, pd.DataFrame([xite_tonic_electric["accuracy"].mean(), xite_tonic_electric["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_tonic_electric"})])
df = pd.concat([df, pd.DataFrame([xite_tonic_heat["accuracy"].mean(), xite_tonic_heat["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_tonic_heat"})])
df = pd.concat([df, pd.DataFrame([xite_phasic_electric["accuracy"].mean(), xite_phasic_electric["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_phasic_electric"})])
df = pd.concat([df, pd.DataFrame([xite_phasic_heat["accuracy"].mean(), xite_phasic_heat["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_phasic_heat"})])
df = pd.concat([df, pd.DataFrame([xite_strong["accuracy"].mean(), xite_strong["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_strong"})])
df = pd.concat([df, pd.DataFrame([xite_strong_heat["accuracy"].mean(), xite_strong_heat["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_strong_heat"})])
df = pd.concat([df, pd.DataFrame([xite_strong_electric["accuracy"].mean(), xite_strong_electric["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_strong_electric"})])
df
```

# 10 Split Crossvalidation

```python
path_xite_all = "xite_all_cross_val_10.csv"
path_xite_heat = "xite_heat_cross_val_10.csv"
path_xite_electric = "xite_electric_cross_val_10.csv"


path_xite_all_embedding = "xite_all_cross_val_10_embedding.csv"
path_xite_heat_embedding = "xite_heat_cross_val_10_embedding.csv"
path_xite_electric_embedding = "xite_electric_cross_val_10_embedding.csv"
```

```python
xite_all = pd.read_csv(path_xite_all)
xite_heat = pd.read_csv(path_xite_heat)
xite_electric = pd.read_csv(path_xite_electric)
xite_all_embedding = pd.read_csv(path_xite_all_embedding)
xite_heat_embedding = pd.read_csv(path_xite_heat_embedding)
xite_electric_embedding = pd.read_csv(path_xite_electric_embedding)
```

```python
df = pd.DataFrame([])
df = pd.concat([df, pd.DataFrame([xite_all["accuracy"].mean(), xite_all["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite all"})])
df = pd.concat([df, pd.DataFrame([xite_heat["accuracy"].mean(), xite_heat["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_heat"})])
df = pd.concat([df, pd.DataFrame([xite_electric["accuracy"].mean(), xite_electric["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite_electric"})])
df = pd.concat([df, pd.DataFrame([xite_all_embedding["accuracy"].mean(), xite_all_embedding["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite all embedding"})])
df = pd.concat([df, pd.DataFrame([xite_heat_embedding["accuracy"].mean(), xite_heat_embedding["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite heat embedding"})])
df = pd.concat([df, pd.DataFrame([xite_electric_embedding["accuracy"].mean(), xite_electric_embedding["accuracy"].std()]).T.rename(columns={0: "mean", 1: "std"}, index={0: "xite electric embedding"})])
df
```

```python
xite_all["accuracy"].plot.kde()
xite_heat["accuracy"].plot.kde()
xite_electric["accuracy"].plot.kde()
plt.legend(["all", "heat", "electric"])
plt.title("Accuracy Classification XITE")
plt.show()
```
